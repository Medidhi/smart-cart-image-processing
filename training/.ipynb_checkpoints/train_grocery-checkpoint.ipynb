# Train the grocery YOLO model

Interactive companion to the `training/` module. It reuses the same `config.yaml`,
`data.yaml`, and `MODELS` registry as `train.py`, so anything you do here matches the
command-line workflow.

**Flow:** setup → pick model → verify dataset → train → review → validate → test → export.

> Run cells top-to-bottom. Switching architecture is one line in **2. Pick the model**
> (or edit `config.yaml`). See `README.md` for the full checklist and `CLAUDE.md` for
> conventions.

## 1. Setup

Install deps into your venv (once) and confirm the training stack + hardware.
On Apple Silicon training uses `mps`; with an NVIDIA GPU it uses `cuda`; otherwise CPU (slow).

```python
# Uncomment the first run to install into the active kernel's environment.
# %pip install -r requirements.txt

import torch
import ultralytics

if torch.cuda.is_available():
    DEVICE = "0"  # first CUDA GPU
elif torch.backends.mps.is_available():
    DEVICE = "mps"  # Apple Silicon
else:
    DEVICE = "cpu"

print(f"ultralytics {ultralytics.__version__}")
print(f"torch       {torch.__version__}")
print(f"device      {DEVICE}")
if DEVICE == "cpu":
    print("WARNING: no GPU/MPS detected — training will be slow.")
```

## 2. Pick the model

Loads `config.yaml` and the `MODELS` registry from `train.py` (single source of truth).
**Change `MODEL` below to switch architectures** — e.g. `yolov8n` → `yolov8s`. Any key in
`MODELS` is valid. Leave `MODEL = None` to use whatever is set in `config.yaml`.

```python
from pathlib import Path

from train import HERE, MODELS, load_config, resolve_weights

# ---- edit me to switch architectures (or None to use config.yaml) ----
MODEL = "yolov8n"
# ----------------------------------------------------------------------

cfg = load_config(HERE / "config.yaml")
if MODEL:
    cfg["model"] = MODEL

model_key = cfg.get("model", "yolov8n")
weights = resolve_weights(model_key)

DATA = (HERE / cfg.get("data", "data.yaml")).resolve()
PROJECT = (HERE / cfg.get("project", "runs")).resolve()

print(f"available models : {', '.join(MODELS)}")
print(f"selected model   : {model_key}  ({weights})")
print(f"dataset spec     : {DATA}")
print(f"epochs={cfg.get('epochs')} imgsz={cfg.get('imgsz')} batch={cfg.get('batch')}")
```

## 3. Verify the dataset

The #1 cause of a bad model is bad data. This validates the layout in `data.yaml`,
prints the class names and image counts, and flags missing/empty splits **before** you
spend time training.

```python
from ultralytics.data.utils import check_det_dataset

info = check_det_dataset(str(DATA))
names = info["names"]
print(f"{len(names)} classes: {list(names.values())}\n")

for split in ("train", "val", "test"):
    p = info.get(split)
    if not p:
        print(f"{split:5}: (not set)")
        continue
    n = sum(1 for _ in Path(p).rglob("*") if _.suffix.lower() in {".jpg", ".jpeg", ".png"})
    print(f"{split:5}: {n} images  ({p})")

print("\nOK — dataset spec resolves. If counts look wrong, fix data.yaml paths/labels.")
```

### Preview a few labeled samples

Draw ground-truth boxes on a handful of training images. Boxes should sit tightly on the
right objects — if they're shifted or mislabeled, your labels are wrong (fix before training).

```python
import cv2
import matplotlib.pyplot as plt

train_dir = Path(info["train"])
img_paths = [p for p in sorted(train_dir.rglob("*"))
             if p.suffix.lower() in {".jpg", ".jpeg", ".png"}][:6]

def label_path(img_path):
    # YOLO layout: .../images/.../x.jpg  <->  .../labels/.../x.txt
    parts = list(img_path.parts)
    if "images" in parts:
        parts[len(parts) - 1 - parts[::-1].index("images")] = "labels"
    return Path(*parts).with_suffix(".txt")

if not img_paths:
    print("No training images found yet — add data under datasets/ then rerun.")
else:
    cols = 3
    rows = (len(img_paths) + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols, figsize=(14, 5 * rows))
    for ax, ip in zip(axes.ravel(), img_paths):
        img = cv2.cvtColor(cv2.imread(str(ip)), cv2.COLOR_BGR2RGB)
        h, w = img.shape[:2]
        lp = label_path(ip)
        if lp.exists():
            for line in lp.read_text().strip().splitlines():
                c, xc, yc, bw, bh = (float(v) for v in line.split()[:5])
                x1, y1 = int((xc - bw / 2) * w), int((yc - bh / 2) * h)
                x2, y2 = int((xc + bw / 2) * w), int((yc + bh / 2) * h)
                cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(img, names[int(c)], (x1, max(12, y1 - 4)),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
        ax.imshow(img); ax.set_title(ip.name, fontsize=8); ax.axis("off")
    for ax in axes.ravel()[len(img_paths):]:
        ax.axis("off")
    plt.tight_layout(); plt.show()
```

## 4. Train

Fine-tunes the selected model using the values from `config.yaml`. Watch the live table:
box/cls/dfl losses should trend **down** and val `mAP50`/`mAP50-95` **up**. If you hit a
CUDA OOM, lower `batch` in `config.yaml` (16 → 8 → 4) and rerun.

```python
from ultralytics import YOLO

model = YOLO(weights)
results = model.train(
    data=str(DATA),
    epochs=cfg.get("epochs", 100),
    imgsz=cfg.get("imgsz", 640),
    batch=cfg.get("batch", 16),
    patience=cfg.get("patience", 25),
    seed=cfg.get("seed", 0),
    device=cfg.get("device") or DEVICE,
    workers=cfg.get("workers", 8),
    project=str(PROJECT),
    name=cfg.get("name", "grocery"),
)

save_dir = Path(results.save_dir)
best = save_dir / "weights" / "best.pt"
print(f"\nbest weights: {best}")
print(f"run dir     : {save_dir}")
```

## 5. Review training results

Displays the curves, confusion matrix, and sample validation predictions written to the
run directory. Look for: converging losses, rising mAP, and which classes get confused.

```python
from IPython.display import Image, display

for fname in ["results.png", "confusion_matrix.png",
              "val_batch0_pred.jpg", "val_batch0_labels.jpg"]:
    fp = save_dir / fname
    if fp.exists():
        print(fname)
        display(Image(filename=str(fp)))
    else:
        print(f"(missing) {fname}")
```

## 6. Validate (metrics)

Re-runs validation on `best.pt` and prints per-class metrics. Rough guide: `mAP50 ≥ 0.7`
is usable; weak classes usually need more/better examples, not just more epochs.

```python
best_model = YOLO(str(best))
metrics = best_model.val(data=str(DATA), device=cfg.get("device") or DEVICE)

print(f"\nmAP50-95: {metrics.box.map:.3f}")
print(f"mAP50   : {metrics.box.map50:.3f}\n")
print(f"{'class':<15}{'mAP50':>8}{'mAP50-95':>10}")
for i, c in enumerate(metrics.box.ap_class_index):
    print(f"{names[int(c)]:<15}{metrics.box.ap50[i]:>8.3f}{metrics.box.ap[i]:>10.3f}")
```

## 7. Quick inference test

Run the trained model on a held-out image and view the predictions. Point `TEST_IMAGE` at
any image path (defaults to the first val image).

```python
val_dir = Path(info["val"])
val_imgs = [p for p in sorted(val_dir.rglob("*"))
            if p.suffix.lower() in {".jpg", ".jpeg", ".png"}]

TEST_IMAGE = str(val_imgs[0]) if val_imgs else None  # or set your own path

if TEST_IMAGE:
    preds = best_model.predict(TEST_IMAGE, conf=0.25, device=cfg.get("device") or DEVICE)
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(preds[0].plot(), cv2.COLOR_BGR2RGB))
    plt.axis("off"); plt.title(Path(TEST_IMAGE).name); plt.show()
else:
    print("No val images to test on — set TEST_IMAGE to a file path.")
```

## 8. Export to ONNX (toward Hailo)

Exports `best.pt` → ONNX with Hailo-friendly settings (opset 11, fixed 640, NMS off, so
`pi/detector.py` needs no changes). Then compile the ONNX to a `.hef` with the Hailo
Dataflow Compiler on Hailo's x86 SDK and drop it into `pi/models/`.

```python
onnx_path = best_model.export(
    format="onnx",
    imgsz=cfg.get("imgsz", 640),
    opset=11,
    dynamic=False,
    simplify=True,
    nms=False,
)
print(f"ONNX written: {onnx_path}")
print("\nNext (on the Hailo x86 SDK, not here):")
print("  hailomz compile --ckpt <this.onnx> --hw-arch hailo8 \\")
print("      --calib-path <images/> --classes", len(names))
print("Then: cp <model>.hef ../pi/models/  and run  python3 server.py --hef models/<model>.hef")
```
